This script treat the raw chamber data of $NO_x$, $NO_y$, and $O_3$. It extract useful data from raw data file. It will give ambient concentration, chamber concentration, and chamber $O_3$ sensitivity.

```python
import os, importlib, sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression

# local module
sys.path.append('../../code/src')
#importlib.reload(sys.modules['extract_chamber_data'])
from extract_chamber_data import get_combined_ambient_data
from extract_chamber_data import get_combined_chamber_data
from extract_chamber_data import get_chamber_averaged_data
from extract_chamber_data import get_o3_sensitivity
from extract_chamber_data import list_of_measurement_day
from extract_chamber_data import get_NOxNOyO3_filepath
from extract_chamber_data import get_raw_lvm
from extract_chamber_data import get_raw_O3
```

## Get $O_3$ sensitivity results for Redlands

```python
date_redlands = list_of_measurement_day("Redlands", test_type='test')
```

```python
df_chamber_result_redlands = pd.DataFrame()
for date in date_redlands:
    
    print('Processing ', date)
    df = get_o3_sensitivity(date, 'Redlands', plot=True)
    df_chamber_result_redlands = df_chamber_result_redlands.append(df)
```

```python
df_chamber_result_redlands.to_csv('../../data/final/O3_sensitivity_redlands.csv')
```

## Get $O_3$ sensitivity for Pasadena

```python
date_pasadena = list_of_measurement_day("Pasadena", test_type='test')
```

```python
#df_chamber_result_pasadena = pd.DataFrame()
for date in date_pasadena:
    if pd.to_datetime(date) > pd.to_datetime('2021-09-14'):
        print('Processing ', date)
        df = get_o3_sensitivity(date, 'Pasadena', plot=True)
        df_chamber_result_pasadena = df_chamber_result_pasadena.append(df)
```

```python
df_chamber_result_pasadena.to_csv('../../data/final/O3_sensitivity_pasadena.csv')
```

## Check $O_3$ sensitivity results by comparing with old vrsion

```python
# prepare df for plotting
df_pasadena_old = pd.read_csv('../../data/raw/Chamber/Summary_O3_PA.csv')
df_pasadena_old['Date'] = pd.to_datetime(df_pasadena_old['Date'])

df_redlands_old = pd.read_csv('../../data/raw/Chamber/Summary_O3_RL.csv')
df_redlands_old['Date'] = pd.to_datetime(df_redlands_old['Date'])

df_chamber_result_redlands['Date'] = pd.to_datetime(df_chamber_result_redlands['Date'])
df_chamber_result_pasadena['Date'] = pd.to_datetime(df_chamber_result_pasadena['Date'])
```

```python
# Redlands
fig, ax=plt.subplots(3,1, figsize=(8,8))

list_para = ['bag2_3hr','b1_b2_3hr', 'b3_b2_3hr']
for i in range(3):
    para = list_para[i]
    ax[i].scatter(df_redlands_old['Date'], df_redlands_old[para])
    ax[i].plot(df_redlands_old['Date'], df_redlands_old[para], label='old')
    ax[i].scatter(df_chamber_result_redlands['Date'], 
                  df_chamber_result_redlands[para])
    ax[i].plot(df_chamber_result_redlands['Date'], 
               df_chamber_result_redlands[para], label='new')
    ax[i].legend()
    ax[i].set_ylabel(para)
ax[2].set_xlabel('Date')
ax[0].set_title('Redlands', fontweight='bold')
```

```python
# Pasadena
fig, ax=plt.subplots(3,1, figsize=(8,8))

list_para = ['bag2_3hr','b1_b2_3hr', 'b3_b2_3hr']
for i in range(3):
    para = list_para[i]
    ax[i].scatter(df_pasadena_old['Date'], df_pasadena_old[para])
    ax[i].plot(df_pasadena_old['Date'], df_pasadena_old[para], label='old')
    ax[i].scatter(df_chamber_result_pasadena['Date'], 
                  df_chamber_result_pasadena[para])
    ax[i].plot(df_chamber_result_pasadena['Date'], 
               df_chamber_result_pasadena[para], label='new')
    ax[i].legend()
    ax[i].set_ylabel(para)
ax[2].set_xlabel('Date')
ax[0].set_title('Pasadena', fontweight='bold')
```

```python
df_chamber_redlands = pd.DataFrame()
for date in date_redlands:
    df = get_combined_chamber_data(date, 'Redlands')
    df_avg = get_chamber_averaged_data(df)
    df_avg['Date'] = pd.to_datetime(date)
    df_chamber_redlands = df_chamber_redlands.append(df_avg)
```

```python
df_chamber_pasadena.to_csv('../../data/final/chamber_measurement_10min_avg_pasadena.csv', 
                           index=False)
df_chamber_redlands.to_csv('../../data/final/chamber_measurement_10min_avg_redlands.csv',
                           index=False)
```
