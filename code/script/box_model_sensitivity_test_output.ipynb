```python
import os, importlib, sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append('../../code/src')
#importlib.reload(sys.modules['extract_model_data'])
from extract_model_data import get_model_O3_sensitivity
```

## Sensitivity test for Pasadena 

```python
# original box model output
filepath_basecase = '../../data/final/Chamber_model_result_sensitivity_test/pasadena_4km_2021_basecase_nopreturbed.out'
filepath_NOx_addition = '../../data/final/Chamber_model_result_sensitivity_test/pasadena_4km_2021_NOx-addition_nopreturbed.out'
df_model_pasadena = get_model_O3_sensitivity(filepath_basecase,
                                                       filepath_NOx_addition,
                                                       NOx_add=1)

# preturbed box model output
filepath_basecase = '../../data/final/Chamber_model_result_sensitivity_test/pasadena_4km_2021_basecase_preturbed_5%.out'
filepath_NOx_addition = '../../data/final/Chamber_model_result_sensitivity_test/pasadena_4km_2021_NOx-addition_preturbed_5%.out'
df_model_pasadena_preturbed5 = get_model_O3_sensitivity(filepath_basecase,
                                                       filepath_NOx_addition,
                                                       NOx_add=1)

# preturbed box model output
filepath_basecase = '../../data/final/Chamber_model_result_sensitivity_test/pasadena_4km_2021_basecase_preturbed_30%.out'
filepath_NOx_addition = '../../data/final/Chamber_model_result_sensitivity_test/pasadena_4km_2021_NOx-addition_preturbed_30%.out'
df_model_pasadena_preturbed30 = get_model_O3_sensitivity(filepath_basecase,
                                                       filepath_NOx_addition,
                                                       NOx_add=1)

filepath_basecase = '../../data/final/Chamber_model_result_sensitivity_test/pasadena_4km_2021_basecase_preturbed200.out'
filepath_NOx_addition = '../../data/final/Chamber_model_result_sensitivity_test/pasadena_4km_2021_NOx-addition_preturbed200.out'
df_model_pasadena_preturbed200 = get_model_O3_sensitivity(filepath_basecase,
                                                       filepath_NOx_addition,
                                                       NOx_add=1)
```

```python
fig, ax = plt.subplots(figsize=(6,3))
ax.plot(df_model_pasadena['Date'], df_model_pasadena['O3_final'], label='orig')
ax.plot(df_model_pasadena_preturbed5['Date'],
        df_model_pasadena_preturbed5['O3_final'],
        label='preturbed +5%')
ax.plot(df_model_pasadena_preturbed30['Date'],
        df_model_pasadena_preturbed30['O3_final'],
        label='preturbed +30%')
ax.plot(df_model_pasadena_preturbed200['Date'],
        df_model_pasadena_preturbed200['O3_final'],
        label='preturbed +200%')
ax.set_ylabel('Final $O_3$ concentration (ppbv)')
ax.set_xlabel('Date')
ax.legend(loc='center', bbox_to_anchor=(1.2, 0.5))
plt.xticks(rotation=30)

# add notation of % of change
change_mean = ((df_model_pasadena_preturbed5['O3_final'] - df_model_pasadena['O3_final'])/
               df_model_pasadena['O3_final']).mean()
change_mean2 = ((df_model_pasadena_preturbed30['O3_final'] - df_model_pasadena['O3_final'])/
               df_model_pasadena['O3_final']).mean()
change_mean3 = ((df_model_pasadena_preturbed200['O3_final'] - df_model_pasadena['O3_final'])/df_model_pasadena['O3_final']).mean()
                
ax.text(0.01, 1.02, f'Average changes (+5%): {change_mean:.2%}\nAverage changes (+30%): {change_mean2:.2%}\nAverage changes (+200%): {change_mean3:.2%}', transform=ax.transAxes, fontsize=10)
ax.set_ylim(0,200)

plt.show()
fig.savefig('../../result/box_model_sensitivity_test_O3_pasadena.png', 
            dpi=600, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(6,3))
ax.plot(df_model_pasadena['Date'], df_model_pasadena['delta_O3'], label='orig')
ax.plot(df_model_pasadena_preturbed5['Date'],
        df_model_pasadena_preturbed5['delta_O3'],
        label='preturbed +5%')
ax.plot(df_model_pasadena_preturbed30['Date'],
        df_model_pasadena_preturbed30['delta_O3'],
        label='preturbed +30%')
ax.plot(df_model_pasadena_preturbed200['Date'],
        df_model_pasadena_preturbed200['delta_O3'],
        label='preturbed +200%')
ax.set_ylabel('$\Delta O_3^{+NO_x}$ (ppbv)')
ax.set_xlabel('Date')
ax.legend(loc='center', bbox_to_anchor=(1.2, 0.5))
plt.xticks(rotation=30)

# add notation of % of change
change_mean = ((df_model_pasadena_preturbed5['delta_O3'] - df_model_pasadena['delta_O3'])/
               df_model_pasadena['delta_O3']).mean()
change_mean2 = ((df_model_pasadena_preturbed30['delta_O3'] - df_model_pasadena['delta_O3'])/
               df_model_pasadena['delta_O3']).mean()
change_mean3 = ((df_model_pasadena_preturbed200['delta_O3'] - df_model_pasadena['delta_O3'])/df_model_pasadena['delta_O3']).mean()
                
ax.text(0.01, 1.02, f'Average changes (+5%): {change_mean:.2%}\nAverage changes (+30%): {change_mean2:.2%}\nAverage changes (+200%): {change_mean3:.2%}', transform=ax.transAxes, fontsize=10)

plt.show()
fig.savefig('../../result/box_model_sensitivity_test_deltaO3_pasadena.png', dpi=600, bbox_inches='tight')
```
